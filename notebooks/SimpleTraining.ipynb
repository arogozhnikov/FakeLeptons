# In this notebook, we try simple classifiers 
The aim to distinguish prompt leptons (those coming from W boson) from others.

The data is decays with $t \bar{t}$ production

```python
import root_numpy
import hep_ml
import pandas
from hep_ml.commonutils import train_test_split
```

```python
folder = '/mnt/w76/notebook/datasets/fromLesya/ver3/'
prompt_filename = folder + 'tt_prompts.root'
fakes_filename  = folder + 'tt_fakes.root'

treename = "fakeTree"
```

## Listing all available branches

```python
all_columns = root_numpy.list_branches(prompt_filename, treename='fakeTree')
sorted(all_columns)
```

## Those features we never use in training

```python
markers = ['_mom', 'gen', '_lep', 'AllMC', '_closeJetPtAllstatus', 
           '_eventNb', '_runNb', '_lumiBlock', '_origin', '_originReduced', '_isolationMC']
ignored_columns = [col for col in all_columns if any([marker in col for marker in markers])]
sorted(ignored_columns)
```

# Preparing information about columns to read

```python
flattened_columns = dict()
for column in all_columns:
    # print column, data
    data = root_numpy.root2array(prompt_filename, treename=treename, branches=[column], stop=10)
    try:
        for index in range(len(data[0][0])):
            flattened_columns["{column}[{index}]".format(column=column, index=index)] = column + str(index)
    except:
        flattened_columns[column] = column
# for some strange reason this thing is also added
flattened_columns.pop('_PVerr[3]')
```

```python
prompt_test = root_numpy.root2array(prompt_filename, treename=treename, branches=sorted(flattened_columns.keys()), stop=1000)
prompt_test = pandas.DataFrame(prompt_test)
```

# Reading the data

```python
# number of events we will use
n_events = 200000
```

```python
read_columns = sorted(flattened_columns.keys())
```

```python
prompt_data = root_numpy.root2array(prompt_filename, treename=treename, branches=read_columns, stop=n_events)
prompt_data = pandas.DataFrame(prompt_data)

fake_data = root_numpy.root2array(fakes_filename, treename=treename, branches=read_columns, stop=n_events)
fake_data = pandas.DataFrame(fake_data)
```

# Plotting the distributions

```python
columns = prompt_data.columns
figure(figsize=[18, 100])
for i, column in enumerate(columns, 1):
    subplot((len(columns) + 2) // 3, 3, i)
    concatenated = numpy.concatenate([prompt_data[column], fake_data[column]])
    limits = numpy.percentile(concatenated, [1, 99])
    hist(prompt_data[column], bins=30, label='prompt', normed=True, range=limits)
    hist(fake_data[column],   bins=30, label='fake',   normed=True, range=limits)
    legend(loc='best')
    title(column)
```

# Features we use in training

```python
train_columns = [col for col in read_columns if not any([marker in col for marker in markers])]
sorted(train_columns)
```

```python
data = pandas.concat([prompt_data, fake_data], ignore_index=True)
data = data[train_columns]
# Three ip features contain the events with huge values not convertible to float32 
data = numpy.clip(data, -1e20, 1e20)


labels = numpy.array([1] * len(prompt_data) + [0] * len(fake_data))

trainX, testX, trainY, testY = train_test_split(data, labels, train_size=0.5)
```

# Training classifiers 

```python
from sklearn.ensemble import GradientBoostingClassifier
# from hep_ml.ugradientboosting import uGradientBoostingClassifier
# from hep_ml.experiments.fastgb import FastGradientBoostingClassifier
# from hep_ml.experiments.gradient_boosting import GradientBoosting
```

```python
classifiers = hep_ml.ClassifiersDict()
classifiers['gb'] = GradientBoostingClassifier(subsample=0.1, n_estimators=100, max_depth=6, 
                                               min_samples_split=50, max_features=8)
```

```python
classifiers.fit(trainX, trainY)
predictions = classifiers.test_on(testX, testY)
```
